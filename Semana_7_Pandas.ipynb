```python
import numpy as np
import pandas as pd
```

```python
rng = np.random.default_rng(12345)
size = 100
id = np.arange(size)
idade = rng.integers(low = 18, high = 120, size = size)
sexo = rng.choice(["F", "M"], size = size)
estado_civil = rng.choice(["solteiro", "casado", "divorciado", "viuvo"], size = size)
sinistros = rng.poisson(2, size = size)
premio = np.exp(rng.normal(5, 3, size))

dados = {
    "id": id,
    "idade": idade,
    "sexo": sexo,
    "estado_civil": estado_civil,
    "sinistros": sinistros,
    "premio" : premio
}

dados = pd.DataFrame(dados)
dados.head()
```

```python
dados["categoria"] = pd.cut(dados["idade"], [18, 30, 45, np.inf], labels = ["Jovem", "Adulto", "Senior"], include_lowest = True)
```

```python
descontos = {
    "Adulto": 0.1,
    "Senior": 0.1,
    "Jovem": 0.05
}

dados["desconto"] = dados["categoria"].map(descontos) * dados["premio"] * (dados["sinistros"] == 0)
```
