Removendo colunas que não serão usadas.

```python
cursos_pd = cursos_pd.drop(
    columns=[
        "ementa",
        "chamadas",
        "publico_alvo",
        "video_1a_aula",
        "instrutores",
        "musicas",
        "depoimentos",
        "transcriptions",
        "highlightedExtraField",
        "curso_substituto",
        "meta_title",
        "video_1a_aula",
        "parceiros",
        "formacoes",
        "instrutores",
        "carreiras",
        "musicas",
        "depoimentos",
        "formacoes",
        "curso_substituto",
        "metadescription",
        "requerimentos",
    ]
)
```

```python
cursos_pd["alunos_ativos_total"] = cursos_pd.alunos_ativos.apply(lambda x: len(x["alunos"]) if x else 0)
```

```python
cursos_pd["nome_categoria"] = cursos_pd.categoria.apply(lambda x: x["nome"])
```

```python
cursos_pd["nome_subcategoria"] = cursos_pd.subcategoria.apply(lambda x: x["nome"])
```

```python
import plotly.express as px

px.histogram(cursos_pd, x="quantidade_alunos")
```

```python
from numpy import inf


q50 = cursos_pd.quantidade_alunos.quantile(0.50)
q75 = cursos_pd.quantidade_alunos.quantile(0.75)

cursos_pd["popularidade"] = pd.cut(cursos_pd.quantidade_alunos, bins=[0, q50, q75, inf], labels=["Moderadamente Popular", "Popular", "Muito Popular"])
cursos_pd[["nome", "quantidade_alunos", "popularidade"]].sort_values(by="quantidade_alunos", ascending=False).head(10)
```

```python
cursos_pd.query("nota_disponivel").groupby("nome_categoria").agg(
    **{
        "Média": pd.NamedAgg(column="nota", aggfunc="mean"),
        "Mínimo": pd.NamedAgg(column="nota", aggfunc="min"),
        "Máximo": pd.NamedAgg(column="nota", aggfunc="max"),
        "Desvio padrao": pd.NamedAgg(column="nota", aggfunc="std"),
    }
).round(1).sort_values(by=["Média", "Máximo"])
```

```python
cursos_pd.groupby(["nome_categoria", "nome_subcategoria"]).agg(
    **{"Quantidade de cursos": pd.NamedAgg(column="id", aggfunc="count")}
).assign(**{"Percentual": lambda x: x["Quantidade de cursos"] / x["Quantidade de cursos"].sum() * 100}).round(2)
```
